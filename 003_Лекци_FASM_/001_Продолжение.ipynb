# С чего начать?
Очень большой проблемой является поиск актуальной литературы на тему "Программирование на Ассемблере".<br/>
В рамках этого вопроса рекомендую посетить следующие сайты...<br/>
Изучать будем FASM.<br/>
# <a href="https://habr.com/ru/post/345748/">Как писать на ассемблере в 2018 году.</a>
В этой статье приводится соблазнительно большой список литературы, к тому же статья может претендавать на краткий обзор вопроса.<br/>
В добавок, на основании этой статьи, можно сказать что приличных пакетов под Ассемблер осталось раз два и обчелся...<br/>
- **MASM** - много версий, документация запутана, требует установки, лицензия... Нередко можно встретить источник литературы по уже несуществующией версии, так что "Удачи"...<br/>
- **TASM** - когда-то был в приличном состоянии. Но сейчас ... он не обновлялся со времен 32 битного Borland, что примерно 2008 год.<br/>
- Следующий в списке - **FASM** - когда-то был темной лошадкой. На сегодняшний момент заслужанно теснит MASM... FASM распространяется свободно, не требует установки, и что удивительно, по сей день активно обновляется. К тому же он неимоверно складно и просто синтаксически устроен. При скачивании с официального сайта идет вместе с примерами. Есть актуальный, недавно обновленный, программстский мануал. Всего этого попрежнему недостаточно. Потому попробуем создать начальное ускорение.<br/>

**По этому предложим свой взгляд на ситуацию...**
- Имеет смысл посетить сайт <a href="https://fasmworld.ru/uchebnik/">FasmWorld.</a> Он посвящен тренировке навыков программирования на FASM под DosBox. Процетирую недословно: "прежде чем браться за Windows, можно поиграться и с DosBox".Из этого этапа изучения Асемблера получается неплохой экскурс в историю. Тем неменее, материал хоть и обладает редким качеством достоверности изложения, на практике половина учащихся неспособна изучить его в том варианте, в котором он есть. Большие объемы теории, резко сваливающиеся на голову, демативируют учащихся. По этой причине будем FasmWorld использовать как дополнительный материал высогоко качества.<br/>
- На одном из форумов был задан вопрос. А сеть ли Руссифицированный мануал программиста под FASM. Ответ был таков... Нет, но подожди, сейчас перегоню через онлайн переводчик. После этого интузиасты, с различной степенью отклонения от исходного текста, с применением личной интерпритации и минимальным объяснением как этим пользоваться стали выкладывать в интернет свои варианты переводов... Я насчитал таковых 2,3 или 4 варианта. Каждый из них не охватывает общую картину, но тем неменее хочется предложить один из наиболее приличных вариантов. <a href="https://www.cyberforum.ru/fasm/thread1240599.html">Мануал программера.flat assembler 1.71</a><br/>

- В тот момент, когда программки в стиле "Привет мир" остались позади, вознивает вопрос: Как со всем этим хозяйством обращаться? Как не тратить время на часто повторяющиеся проблемы?
Именно в этот самый момент в дело вступают макросы. <a href="https://www.cyberforum.ru/fasm/thread1253291.html">Руководство по препроцессору FASM.</a> В программировании 2000 годов в литературе часто астречается антипатерн "GoldenHammer" или золотой молоток, что означает решение всех проблем одним способом. Антипатерны являются обратной стороной медали патернов проектирования. Хоть в литературе и повторяют заученную фразу:"Только патерны и никаких антипатернов" - разумное применение любых приемов ещё никому не вредило!!! В варианте FASM, судя повсему пошли по пути максимального упрощения, и очень талантливо решили почти все проблемы при помощи макросов.<br/>
Да, FASM, только компилятор, не литнкер. Не объединяет заранее скомпилированные файлы. Зачем такие сложности!!!
Проблема решается через препроцессинг, макросы и подшивание $*$.Inc файлов в один на этапе перед подачей на компиляцию.<br/>
Получился поражающей одновременно простой и универсальностью инструментарий. Именно по этой причине можно сказать, что на лицо не каноническое применение антипарена золотой молоток.<br/> Если удастся учащимся освоение макросов, то оно превратит процесс написания программных кодов ассемблера из решения ребуса в увлекательное программирование на близком аналоге  императивного языка высокого уровня. <br/>
Вы удивитеся узнав, что так было не всегда. Капнув чуть глубже интернет форумы можнео узнать что MASM пестрит разнашерстным синтаксисом в различных версиях на тему линкинга, предкомпиляции и сегментации памяти программы. <br/>
Возможно под FASM встречается куда больше примеров работающего программного кода чем под MASM. <br/>
В этом плане на экзамине легко узнать, кто из студентов отлынивал во время лекций. Они находят программный код, тасуют его не глядя и пытаются сдавать... При этом очень сильно удивляются на то что он ещё должен компилироваться без ошибок и работать. <br/> Дело доходит до того что сразу видно какие блоки разменом примерно в 10-15 строк взяты из FASM, а какие из MASM. **Делаем вывод. Программный код ничего не стоит если он не компилируется и  не запускается!!!**<br/>
Именно этому вопросу мы и собираемся удилить особое внимаение!!!<br/>

# Начинаем писать 
Основная проблема юного ассемблериста, это вывод на экран состояний регистров.<br/>
Что бы это провернуть, требуется средний уровень владения языком (не меньше).<br/>
Это первая из преград, которая преодолевается нахождением адекватной среды разработки.<br/>
**EMU8086** - хотелось бы сказать, но за прошедшее десятилетие среда стала платной.<br/>
**Free Pascal 3.0/Lazarus/Delphi** - хотелось бы сказать, но среда не является Mейн Cтримом дня сегодняшнего.<br/>
**Visual Studio 2019/C++** - тоже обладает своими неочевидными недостатками, но пользоваться вы ей будем, хоть и не с самого начала...<br/>
А потому, встречайте: <a href="https://kpolyakov.spb.ru/prog/lamp.htm">ЛамПанель.</a>

Сразу скажу, ЛамПанель ниначто не способна, кроме визуализации... Что на определенном этапе себя оправдывает.<br/>
Есть Русскоязычный хелп на 2 страници А4.<br/>
Есть $*$.PDF на 14 страници.<br/>
Есть  все что требуется для изучения.<br/>
- Задание. Часть первая.

Нарисовать, авторскую картинку.
- Задание. Часть вторая.

Сделать анимацию. Циклический сдвиг с низу в верх, с лева на право.

Студентам можно сказать что есть:
- 4 регистра, r0,r1,r2,r3
- 8 портов для работы с внешними устройствами (имитация), p0,p1,p2,p3,p4,p5,p6,p7.
- in/out - пересылка машинных слов между регистрами и портами.
- mov - команда пересылки между регистрами и оперативной памятью (об этом подробнее позже).
- Есть системные процедуры, которые можно посмотреть как "программма посмотреть ПЗУ".

Подробнее о том как сделать это...<br/>


Создаем файл Excel, делаем эскиз рисунка из нулей и единиц, переводим при помощи таблиц в 16 ричный формат (с лева на право). Осталось подствить в программный код.
