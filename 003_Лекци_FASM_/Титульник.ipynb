Номинация "Инновационные технологии обучения" <br/>
В номинации участвуют методические разработки по применению современного инструментария образовательного процесса (проблемные лекции, семинары, деловые игры кейсы, квесты с использованием компьютерных симуляторов и т.д.).<br/>
Данный материал относится к проблемным лекциям. <br/>

Материалы написан в "Jupiter Notebook"(2015 год появления).<br/>
В качестве ядра используется язык Julia (2018 год появления).<br/>
Julia - язык предназначенный для постановки и проведения вычислительных экспериментов, заточен под высокую производительность.<br/>
Julia имеет возможность написания ассемблерных вставок (в отличае от Pyton).<br/>
Julia - язык высокгого уровня который при этом является "асемблероблизким языком".<br/>
Все это позволяет в последствии перейти к изучению описанного технологиеского стека.<br/>
Но не так все просто.<br/>
Асемблер ещё никому легко не дался, и наша задача это изменить (по возможности).<br/>
"Jupiter Notebook" - позволяет не только писать тексты с HTML форматированием, но и запускать написанные на его ядре программные коды "Прямо из лекций".<br/>
Планируется написать прослойку на Julia, которая позволит запускать коды и других не столь стандартных (как Pyton) языков программирования.<br/>

Все программные коды из этих проблемных лекций либо запускаются из самой лекции (если она не в PDF формате),  либо выложенны на GitHub. При правильном клонировании с GitHub в папку "C:/D" все примеры программных кодов также запускаются... Ссылка на git репозиторий выдается студентам во время лекции.
