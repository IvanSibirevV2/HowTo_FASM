# Знакомство с миром
### Базовые инструменты командной разработки.
<a href="https://techrocks.ru/2019/02/19/git-origin-story/">История происхождения Git.</a><br/>
Краткий пересказ.<br/>
- **CVS** - с 1980 года, одна из первых систем контроля версий, по сравнению с сегодняшними огромное колличество недостатков.
- **BitKeeper/BitMover** - примерно 2000-ные годы, распределенная систем контроля версий. Были проблемы при разработки и выкладывании Linux.
После судебных исков, был написан Git.
- **Git**- июнь 2005 год, система контроля версий, которая превзошла остальные и стала в последующие годы безспорным промышленным стандартом.

<a href="https://git-scm.com/download/win">Downloading Git.</a><br/>
Ставится на компьютер. Работает как консоль.<br/>
Можно писать от руки (от клавиатуры) команды. Можно писать $*$.bat файлы (пакетно запускаемые списки команд).Можно подключаться программно через дочернюю консоль или через Temp.bat, временный файл с удалением после исполнениячерез CMD.
Можно работать с удаленными репозиториями.<br/>
Подробнее об этом в официальной докементации <a href="https://git-scm.com/doc">Git Documentation</a>.<br/>
Натренироваться можно. Вот онлайн JavaScript приложение <a href="https://learngitbranching.js.org/?locale=ru_RU">GitLearn</a>.<br/>
Там достаточно уроков и онлайн упражнений, чтобы затем, чувствовать себя в консоли комфортно.<br/>

Для быстрого вхождения в распределенную систему контроля версий можно использовать сторонние программы, что предоставляют англо/русскоязычный интерфейс. Вот некоторые изних:<br/>

Дело в том, что производители таких сторонних приложений могут не отличаться добросовестностью и подворовывать логины и пароли, прокладывать сетевые дыры и просто по мелкому портить жизнь. Этим пугают обычно новичков на производстве, заставляя учить консоль Git.<br/>
Но а мы, пока что, выбираем свой пользователский интерфейс, которым будет удобнее пользщоваться.<br/>

# Немного  теории.
Git позволяет создавать репозитории: центральный (без рабочего каталога) и персональный (с рабочим каталогом). Центральный репозиторий обычно используется как центральное хранилище и содержит файлы системы контроля версиий Git. Персональный - содержит все тоже самое только в скрытой папке внутри себя. <br/>
Любой репозиторий пользователем компьютера воспринимается как папка. <br/>
Пользователь может осуществлять индексирование файлов, то есть занесение изменений файлов в индекс. Пользователь может осуществлять коммит, то есть сохранения в репозитории состояния проиндексированных файлов.<br/>
